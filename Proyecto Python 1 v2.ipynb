```python
# 1. Función que recibe una cadena de texto como parámetro y devuelve un diccionario con las frecuencias de cada letra en la cadena, sin considerar los espacios.

def contar_letras(texto): #defino la función que va a contar las letras
    conteo = {}           #creo un diccionario vacío para que se guarde la cuenta (frecuencia) de cada letra
    for letra in texto:
        if letra == " ":  #si la letra es un espacio, pasa a la siguiente letra
            continue
        if letra in conteo: #si la letra ya está en el diccionario "conteo", suma 1
            conteo[letra] = conteo[letra] + 1
        else:
            conteo[letra] = 1 #si la letra no está en el diccionario "conteo", la añade con el valor 1
    return conteo
print(contar_letras("estoy haciendo mi primer proyecto de python"))
```

```python
# 2. Dada una lista de números, obtengo una nueva lista con el doble de cada valor utilizando la función map().

def doblar_numero(n):   #creo la función que va a calcular el doble de un número
    return n * 2       
def doblar_lista(numeros):  #creo la función a llamar cuando queramos obtener la nueva lista con el doble de cada número
    resultado_dobles = map(doblar_numero, numeros)  #aplico las funciones map() y "doblar_numero" a cada número de la lista y lo guardo en la variable "resultado_dobles"
    resultado_lista = list(resultado_dobles)  #creo la lista que almacena el resultado de aplicar la función map()
    return resultado_lista

prueba_numeros = [3, 6, 9]
print(doblar_lista(prueba_numeros))
```

```python
# 3. Función que toma una lista de palabras y una palabra objetivo como parámetros.
# La función devuelve una lista con todas las palabras de la lista original que contienen la palabra objetivo.

def encontrar_palabras_con_objetivo(lista_palabras, palabra_objetivo): #creo la función que busca palabras en la lista que contengan la palabra objetivo
    coincidencias = []  #creo una lista vacía para almacenar las palabras de la lista original que coinciden con la palabra objetivo
    for palabra in lista_palabras:  #se recorre cada palabra de la lista original
        if palabra_objetivo in palabra:  #se comprueba si la palabra objetivo está dentro de la palabra actual en el bucle
            coincidencias.append(palabra)  #y si está, se añade a la lista "coincidencias"
    return coincidencias

palabras = ["estudia", "proyecto", "python", "estudiante"]
print(encontrar_palabras_con_objetivo(palabras, "estudia"))
```

```python
# 5. Función que toma una lista de números como parámetro y un valor opcional "nota_aprobado", que por defecto es 5. La función calcula la media de los números en la lista
# y determina si la media es mayor o igual que "nota_aprobado". Si es así, el estado será "aprobado", de lo contrario, será "suspenso". 
#La función devuelve una tupla que contiene la media y el estado.

def evaluacion_notas(notas, nota_aprobado = 5):  #creo la función que calcula la media de las notas y fijo el valor de aprobado en 5
    if len(notas) == 0:    #añado la validación de que la lista no esté vacía, y defino el error que debe aparecer si es el caso
        raise ValueError("La lista de notas está vacía")
    media = sum(notas) / len(notas)   #creo la función que calcula la media y fijo la condición del estado "aprobado" o "suspenso"
    if media >= nota_aprobado:
        estado = "aprobado"
    else:
        estado = "suspenso"
    resultado = (media, estado)
    return resultado

print(evaluacion_notas([3.5,6,8,5.75]))
```

```python
# 6. Función que calcula el factorial de un número de manera recursiva.

def factorial(numero):
    if numero == 0 or numero == 1:  #si el número es 0 o 1, su factorial es 1
        return 1
    else:
        return numero * factorial(numero - 1)  #si el número es distinto de 0 o 1, se multiplica por el factorial del número anterior (n*(n-1)*(n-2)*(n-3)...)
    
print(factorial(5))
```

```python
# 7. Función que convierte una lista de tuplas a una lista de strings, utilizando la función map().

def convertir_a_string(lista_tuplas):  #creo la función que convierte la lista de tuplas en una lista de strings
    def tupla_a_string(tupla):         #creo la función que primero convierte una tupla en un string
        elementos_strings = []         #creo una lista vacía que almacena los strings
        for elemento in tupla:         #recorre la tupla y convierte cada elemento en string
            elementos_strings.append(str(elemento))
        resultado = " ".join(elementos_strings)  #une todos los elementos dejando un espacio entre ellos
        return resultado
    lista_strings = list(map(tupla_a_string, lista_tuplas))  #aplico la función map() y "tubla_a_string" a cada tupla de la lista
    return lista_strings

tuplas = [("calle", "avenida"), ("2","4")]
print(convertir_a_string(tuplas))
```

```python
# 29. Función que convierte una variable en una cadena de texto y enmascara todos los caracteres con el caracter '#', excepto los últimos cuatro.

def enmascarar(variable):
    texto = str(variable)
    if len(texto) <= 4:  #si el texto tiene 4 caracteres o menos, no se enmascara ningún caracter
        return texto
    numeros_enmascarados = len(texto) - 4  #calcula cuántos caracteres debe enmascarar: todos menos los últimos 4
    parte_enmascarada = "#" * numeros_enmascarados  #se crea la parte enmascarada usando '#' 
    ultimos_cuatro = texto[-4:]  #se toman los últimos 4 caracteres del texto dado
    return parte_enmascarada + ultimos_cuatro

print(enmascarar("123456789"))
```

```python
# 34. Creación clase Arbol.

class Arbol:
    def __init__(self):
        self.tronco = 1  #tronco de longitud 1 y lista de ramas vacía
        self.ramas = []

    def crecer_tronco(self):
        self.tronco = self.tronco + 1  #aumenta la longitud del tronco en una unidad

    def nueva_rama(self):
        self.ramas.append(1)

    def crecer_ramas(self):
        for i in range(len(self.ramas)):  #aumenta en una unidad todas las ramas
            self.ramas[i] = self.ramas[i] + 1

    def quitar_rama(self,posicion):
        indice = posicion - 1
        if 0 <= indice < len(self.ramas):
            self.ramas.pop(indice)

    def info_arbol(self): #devuelve información del árbol
        return {
            "longitud_tronco": self.tronco,
            "numero_ramas": len(self.ramas),
            "longitudes_ramas": self.ramas[:]
        }

#CASO DE USO:

arbol = Arbol()             #crear un árbol
arbol.crecer_tronco()       #crecer el tronco en una unidad
arbol.nueva_rama()          #añadir una nueva rama
arbol.crecer_ramas()        #hacer crecer todas las ramas en una unidad
arbol.nueva_rama()          #añadir dos nuevas ramas
arbol.nueva_rama()
arbol.quitar_rama(2)        #retirar la rama situada en la posición 2
print(arbol.info_arbol())   #obtener información del árbol
```

```python
# 36. Creación clase UsuarioBanco.

class UsuarioBanco:  #inicializa un usuario del banco con su nombre, saldo y si tiene o no cuenta corriente
    def __init__(self, nombre, saldo, cuenta_corriente):
        self.nombre = nombre
        self.saldo = saldo
        self.cuenta_corriente = cuenta_corriente

    def retirar_dinero(self, cantidad):  #resta dinero del saldo del usuario. Si no hay suficiente saldo, lanza un error
        if cantidad <= 0:
            raise Exception("La cantidad a retirar debe ser mayor que 0.")
        if cantidad > self.saldo:
            raise Exception(f"{self.nombre} no tiene suficiente saldo para retirar {cantidad}.")
        self.saldo -= cantidad
        print(f"{self.nombre} ha retirado {cantidad}. Saldo actual: {self.saldo}")

    def transferir_dinero(self, otro_usuario, cantidad):
        if not otro_usuario.cuenta_corriente or not self.cuenta_corriente:
            raise Exception("Ambos usuarios deben tener cuenta corriente para realizar una transferencia.")
        if cantidad <= 0:
            raise Exception("La cantidad a transferir debe ser mayor que 0.")
        if cantidad > otro_usuario.saldo:
            raise Exception(f"{self.nombre} no tiene suficiente saldo para transferir {cantidad}.")
    
        self.saldo -= cantidad  #se descuenta del usuario que envía y se suma al receptor
        otro_usuario.saldo += cantidad
        print(f"{self.nombre} ha transferido {cantidad} a {otro_usuario.nombre}.")

    def agregar_dinero(self, cantidad):
        if cantidad <= 0:
            raise Exception("La cantidad a agregar debe ser mayor que 0.")
        self.saldo += cantidad
        print(f"{self.nombre} ha agregado {cantidad}. Saldo actual: {self.saldo}")

#CASO DE USO:

#Creo dos usuarios: Alicia y Bob
alicia = UsuarioBanco("Alicia", 100, True)
bob = UsuarioBanco("Bob", 50, True)    

#Agrego 20 unidades al saldo de Bob
bob.agregar_dinero(20)

#Transfiero 80 unidad desde Bob a Alicia (añado el "try" por validar que el resto del código es correcto, porque el programa se detenía ante el error de saldo insuficiente)
try:
    alicia.transferir_dinero(bob, 80)
except Exception as excepcion:
    print("Error:", excepcion)

#Retiro 50 unidades del saldo de Alicia
alicia.retirar_dinero(50)

#Saldos finales:
print(f"Saldo final de {alicia.nombre}: {alicia.saldo}")
print(f"Saldo final de {bob.nombre}: {bob.saldo}")
```

```python
# 37. Función llamada procesar_texto que procesa un texto según la opción especificada: contar_palabras, reemplazar_palabras, eliminar_palabra.

def contar_palabras(texto):  #función para contar el número de veces de que aparece cada palabra en el texto
    palabras = texto.split() #separa el texto por espacios para obtener una lista de palabras
    conteo = {}              #creo el diccionario vacío que devolverá la función con los resultados
    for palabra in palabras:
        if palabra in conteo:
            conteo[palabra] += 1 #si la palabra ya existe en el diccionario, sumamos 1
        else:
            conteo[palabra] = 1  #si la palabra no existe, la añadimos
    return conteo

def reemplazar_palabras(texto, palabra_original, palabra_nueva):  #función para reemplazar una "palabra_original" del texto por una "palabra_nueva"
    palabras = texto.split() #convierte el texto en lista 
    resultado = []   #creo una lista vacía
    for palabra in palabras:
        if palabra == palabra_original:
            resultado.append(palabra_nueva)  #si la palabra coincide con la original, la reemplazamos por la nueva
        else:
            resultado.append(palabra)        #si no coincide, se mantiene igual
    return " ".join(resultado)               #une la lista de nuevo separando por espacios

def eliminar_palabra(texto, palabra_a_eliminar):  #función que elimina una palabra del texto
    palabras = texto.split()      #convierte el texto en lista
    resultado = []                #creo una lista vacía 
    for palabra in palabras:
        if palabra != palabra_a_eliminar:  #solo se añaden a la lista las palabras distintas a la que queremos eliminar
            resultado.append(palabra)
    return " ".join(resultado)

def procesar_texto(texto, opcion, *args):  #función que decide qué operación realizar según la opción elegida ("contar", "reemplazar", "eliminar")
    if opcion == "contar":                 #si la opción elegida es "contar", llama a la función "contar_palabras"
        return contar_palabras(texto)
    elif opcion == "reemplazar":           #si la opción elegida es "reemplazar", llama a la función "contar_palabras"
        return reemplazar_palabras(texto, *args) #args contiene la palabra original y la nueva
    elif opcion == "eliminar":             #si la opción elegida es "eliminar", llama a la función "contar_palabras"
        return eliminar_palabra(texto, *args)    #args contiene la palabra a eliminar
    else:
        return "La opción elegida no es válida. Usa 'contar', 'reemplazar' o 'eliminar'."

#CASO DE USO:

#Cuenta cuántas veces aparece cada palabra en el texto que proporciono
texto = "Tiene clase en la universidad por la mañana"
print("Resultado de contar palabras:")
print(procesar_texto(texto, "contar"))

#Reemplaza la palabra "mañana" por "tarde"
print("Resultado de reemplazar palabra:")
print(procesar_texto(texto, "reemplazar", "mañana", "tarde"))

#Elimina la palabra "tiene"
print("Resultado de eliminar palabra:")
print(procesar_texto(texto, "eliminar", "Tiene"))
```

```python
# 40. Función que toma dos parámetros: figura (puede ser "rectángulo", "círculo" o "triángulo") y datos (una tupla con los datos necesarios para calcular el área de la figura).

def calculo_area(figura, datos):
    if figura in ("rectángulo", "rectangulo"):   
        if isinstance(datos, tuple) and len(datos) == 2:
            base, altura = datos
            area = base * altura
            return f"El área del rectángulo es: {area}"
        else:
            print("Error: debes aportar 2 valores, para la base y la altura.")
    elif figura in ("triángulo", "triangulo"):
        if isinstance(datos, tuple) and len(datos) == 2:
            base, altura = datos
            area = (base * altura) / 2
            return f"El área del triángulo es: {area}"
        else:
            print("Error: debes aportar 2 valores, para la base y la altura.")  #actualización: creo también avisos de error si no se han aportado el número de valores necesarios para el cálculo del área.
    elif figura in ("círculo", "circulo"):
        if isinstance(datos, int):
            (radio) = datos
            area = 3.14159265 * (radio ** 2)   #añado directamente el valor aproximado de pi
            return f"El área del círculo es: {area}"
        else:
            print("Error: debes aportar un solo valor para el radio.")
    else:
        return "Figura no válida. Debe ser 'rectángulo', 'triángulo' o 'círculo'."
    
print(calculo_area("rectangulo", (5)))
```
